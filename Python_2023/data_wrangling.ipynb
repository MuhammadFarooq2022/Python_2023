## **EDA(Exploratory Data Analysis):**

Exploratory Data Analysis (EDA) is a method used to analyze and summarize datasets. In Python, you can use the Pandas library to perform EDA on a dataset.

```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import dataset
df = sns.load_dataset("titanic")
```

```python
# visualize the nan values in heatmap
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap="viridis")
```

```python
df.drop(columns = "deck", inplace=True)
```

```python
df.columns
```

```python
df.shape
```

```python
# visualize and explore
sns.countplot(data=df, x="survived")
```

```python
sns.barplot(data=df, x="survived", y="fare", hue="who")
```

```python
sns.countplot(x="pclass", data=df)
```

```python
# data distribution
sns.histplot(data=df, x="age", bins=30)
```

```python
df["age"].hist(bins=30, color="darkred", alpha=0.7)
```

```python
sns.boxplot(data=df, x="survived", y="age", hue="who")
```

```python
# scatterplot
sns.scatterplot(x="age", y="fare", data=df, hue="pclass")
plt.ylim(0,270)
```

```python
sns.pairplot(df, hue="survived")
```

```python
sns.pairplot(df, hue="survived", vars=["age", "fare", "pclass"])
```

```python
#stacked barchart/plot
df.groupby(["pclass", "who"])["fare"].mean().unstack().plot(kind="bar", stacked=True)
```

```python
# swarm plot
sns.swarmplot(data=df, x="survived", y="age", hue="who")
```

```python
sns.catplot(x="pclass", y="fare", hue="embarked", kind="point", data=df)
```

```python
sns.lmplot(data=df[df["fare"] < 300], x="age", y="fare", hue="pclass")
```

```python
# relation of two numeric variables
sns.lmplot(data=df, x="age", y="fare", hue="pclass")
plt.ylim(0, 300)
```

```python
# how to add a column and make a age group
df.columns
```

```python
df["age_groups"] = pd.cut(df["age"], [0,10,18,30,50,80],
                          labels=["kids", "child","youth","middle age", "old"] )
```
